#### Analyze the Distribution

Using your sample mean, sample standard deviation, and histogram to help, what is the approximate theoretical distribution of the data?

X ~ _____(_____,_____)

How does the histogram help you arrive at the approximate distribution?


The theoretical distribution of the data can be approximated by looking at the shape of the histogram and the calculated sample mean and standard deviation.

If the histogram is symmetric and bell-shaped, and most of the data is concentrated around the mean, the data might follow a Normal distribution.

In this case, you would write it as:

X ~ N(mean, standard deviation)

If the histogram is skewed to the right (i.e., it has a long tail on the right), the data might follow a Log-Normal or Exponential distribution.

If the histogram is skewed to the left (i.e., it has a long tail on the left), the data might follow a Weibull distribution.

If the histogram has multiple peaks, the data might be multimodal, suggesting a mixture of different distributions.

The histogram helps in approximating the distribution because it provides a visual representation of the data's distribution. By looking at the shape of the histogram, you can get a sense of whether the data is symmetric or skewed, unimodal or multimodal, which are key characteristics when determining the type of theoretical distribution that best fits the data.

To definitively determine the distribution of your data though you would need to perform a goodness-of-fit test, such as the Kolmogorov-Smirnov test, the Anderson-Darling test, or the Chi-Square test. These tests compare your data to a theoretical distribution to see if it's a good fit.

Real-world data often doesn't perfectly follow a theoretical distribution. The goal is to find the distribution that best approximates your data and provides a useful model for analysis or prediction.

Below we use laplace, normal and exponential distribution on our samples.

```python
samples = {
    'Lap 2': [125, 131, 131, 131, 128, 129],
    'Lap 3': [126, 132, 133, 132, 127, 129],
    'Lap 4': [125, 131, 131, 131, 127, 129],
    'Lap 5': [124, 132, 134, 130, 130, 129],
    'Lap 6': [125, 132, 134, 129, 127, 129],
    'Lap 7': [125, 132, 131, 129, 129, 129]
}


distributions = [st.laplace, st.norm, st.expon]
for lap, times in samples.items():
    best_distribution = st.norm
    best_params = (0.0, 1.0)
    best_sse = np.inf
    for distribution in distributions:
        params = distribution.fit(times)
        sse = np.sum(np.power(times - distribution(*params).rvs(len(times)), 2.0))
        if best_sse > sse > 0:
            best_distribution = distribution
            best_params = params
            best_sse = sse

    print(f'Best fit for {lap}: {best_distribution.name} distribution with parameters {best_params}')
```

```python
samples = {
    'Lap 2': [125, 131, 131, 131, 128, 129],
    'Lap 3': [126, 132, 133, 132, 127, 129],
    'Lap 4': [125, 131, 131, 131, 127, 129],
    'Lap 5': [124, 132, 134, 130, 130, 129],
    'Lap 6': [125, 132, 134, 129, 127, 129],
    'Lap 7': [125, 132, 131, 129, 129, 129]
}


all_samples = [time for times in samples.values() for time in times]
mean = np.mean(all_samples)
std_dev = np.std(all_samples)


Q1 = st.norm.ppf(0.25, loc=mean, scale=std_dev)
Q3 = st.norm.ppf(0.75, loc=mean, scale=std_dev)
IQR = Q3 - Q1
P15 = st.norm.ppf(0.15, loc=mean, scale=std_dev)
P85 = st.norm.ppf(0.85, loc=mean, scale=std_dev)
median = st.norm.ppf(0.5, loc=mean, scale=std_dev)
prob_gt_130 = 1 - st.norm.cdf(130, loc=mean, scale=std_dev)

print(f'The IQR goes from {Q1} to {Q3}.')
print(f'IQR = {IQR}.')
print(f'The 15th percentile is {P15}.')
print(f'The 85th percentile is {P85}.')
print(f'The median is {median}.')
print(f'The probability that a randomly chosen lap time is more than 130 seconds is {prob_gt_130}.')
```

According to the Central Limit Theorem (CLT), the distribution of sample means will approach a normal distribution as the sample size increases. The mean of this distribution will be equal to the population mean, and the standard deviation (also known as the standard error) will be equal to the population standard deviation divided by the square root of the sample size.

#### 3. Are the sample means for n = 5 and n = 10 “close” to the theoretical mean, μx? Explain why or why not.

We calculate the means below and they are consistent with the Central Limit Theorem (CLT), which states that the distribution of sample means will be centered around the population mean. The fact that the sample means are equal to the population mean suggests that our samples are representative of our population. This is a good indication that our sampling method is reliable.

```python
X_values = [1, 5, 2, 5, 6, 1, 2, 6, 5, 2, 5, 1, 1, 3, 2, 2, 2, 4, 6, 1, 6, 5, 2, 5, 1, 6, 4, 1, 6, 2, 3, 4, 5, 6, 6, 1, 1, 2, 1, 6, 1, 2, 6, 2, 6, 2, 2, 11, 5, 5, 4, 6, 5, 1, 1, 2, 4, 3, 6, 5]


population_mean = np.mean(X_values)
n = 5
means_n5 = [np.mean(X_values[i:i+n]) for i in range(0, len(X_values), n)]
sample_mean_n5 = np.mean(means_n5)
n = 10
means_n10 = [np.mean(X_values[i:i+n]) for i in range(0, len(X_values), n)]
sample_mean_n10 = np.mean(means_n10)

print(f"Population mean (μx): {population_mean}")
print(f"Sample mean for n=5: {sample_mean_n5}")
print(f"Sample mean for n=10: {sample_mean_n10}")
```

#### 4. Which of the two distributions of sample means has the smaller standard deviation? Why?

The distribution of sample means for n=10 will have a smaller standard deviation than the distribution for n=5. This is a result of the Central Limit Theorem (CLT), which states that the standard deviation of the sampling distribution (known as the standard error) is equal to the standard deviation of the population divided by the square root of the sample size (n).

#### 5. As n changed, why did the shape of the distribution of the data change? Use one to two complete sentences to explain what happened.

As n increases, the shape of the distribution of sample means becomes more normally distributed due to the Central Limit Theorem (CLT). This is why we see a change in the shape of the distribution as n changes.
